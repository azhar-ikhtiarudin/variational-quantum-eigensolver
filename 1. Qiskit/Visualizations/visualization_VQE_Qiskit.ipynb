```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
au_to_kcal_conversion = 627.509474
```

```python
df = pd.read_excel('VQE_Qiskit_Data.xlsx')
# df

plt.rcParams["font.family"] = "Times New Roman"

plt.rcParams.update({'font.size': 14})

plt.plot(df['AerSimulator - shots=2048'][:40], label='Qiskit AerSimulator (shots=2048)', color='navy')
plt.plot(df['AerSimulator - statevector'][:40], label='Qiskit AerSimulator (statevector)', color='green')
plt.plot(df['IBM Quantum Osaka'], label='IBM Quantum Osaka', color='darkorange')
plt.axhline(-1.13709, label='Referensi Eksak', linestyle='--', color='maroon')

plt.ylabel("$E_{GS}$  (Hartree)")
plt.xlabel("Jumlah Iterasi")
plt.legend()
```
