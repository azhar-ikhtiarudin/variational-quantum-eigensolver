```python
def calc_vqe_N2H4(i, active_electrons, active_orbitals, max_iter):
    symbols = ["N", "H", "H", "N", "H", "H"]
    geometry = coordinates[i]
    print(geometry)

    H, qubits = qchem.molecular_hamiltonian(
        symbols,
        geometry,
        active_electrons=active_electrons,
        active_orbitals=active_orbitals,
        method="pyscf"
    )
    dev = qml.device("lightning.qubit", wires=qubits)

    # Hartree-Fock state
    hf_state = qchem.hf_state(active_electrons, qubits)
    # generate single- and double-excitations
    singles, doubles = qchem.excitations(active_electrons, qubits)
    def circuit(params, wires):
        qml.templates.AllSinglesDoubles(params, wires, hf_state, singles, doubles)

    wires = dev.wires.tolist()


    @qml.qnode(dev)
    def energy_expval(params):
        circuit(params, wires)
        return qml.expval(H)
    
    # opt = qml.RMSPropOptimizer(stepsize=0.2)
    # opt = qml.AdamOptimizer()
    opt = qml.GradientDescentOptimizer(stepsize=0.8)
    energies = []

    np.random.seed(1967)
    params = np.random.normal(0, np.pi, len(singles) + len(doubles))

    start = time.time()
    for i in range(max_iter):
        params = opt.step(energy_expval, params)

        E = energy_expval(params)

        energies.append(E)

        # log_metric(metric_name="energy", value=e, iteration_number=i)

        print(f"Completed iteration {i + 1}")
        print("Energy:", E)
        print("----------------")

    print(f"Optimized energy: {E} Ha")
    print(f"Elapsed: {time.time()-start} s")

    return energies
```

```python
config = [[6, 6]]
max_iter = 50

for i in range(len(config)):
    print("Configuration: ", max_iter, config[i][0], config[i][1])
    energies = calc_vqe_N2H4(0, active_electrons=config[i][0], active_orbitals=config[i][1], max_iter=max_iter)
    
    file_path = f"data/N2H4_{max_iter}_{config[i][0]}_{config[i][1]}.txt"
    # Open the file in write mode
    with open(file_path, "w") as file:
        # Write each element of the array to a new line
        for element in energies:
            file.write(str(element) + "\n")
```
