```python
import numpy as np
import matplotlib.pyplot as plt
au_to_kcal_conversion = 627.509474
```

```python
work_range = 13
r_array = np.loadtxt('standard_vqe_data_LiH.txt', delimiter=',', skiprows=1)[:,0][:work_range]
standard_vqe_data = np.loadtxt('standard_vqe_data_LiH.txt', delimiter=',', skiprows=1)[:,1][:work_range]
adapt_vqe_data_2 = np.loadtxt('LiH_adapt_vqe_data_e_1.txt', delimiter=',', skiprows=1)[:,1][:work_range]
adapt_vqe_data_4 = np.loadtxt('LiH_adapt_vqe_data_e_1_active_electron_4.txt', delimiter=',', skiprows=1)[:,1][:work_range]
adapt_vqe_data_4_e3 = np.loadtxt('LiH_adapt_vqe_data_e_3_active_electron_4.txt', delimiter=',', skiprows=1)[:,1][:work_range]
fci_data = np.loadtxt('fci_data_LiH.txt', delimiter=',', skiprows=1)[:,1][:work_range]
hf_data = np.loadtxt('hf_data_LiH.txt', delimiter=',', skiprows=1)[:,1][:work_range]
```

```python
# plt.plot(r_array, adapt_vqe_data_2, label='ADAPT-VQE e 2', marker='o', markersize=5)
# plt.plot(r_array, adapt_vqe_data_4, label='ADAPT-VQE e 4', marker='o', markersize=5)

plt.plot(r_array, hf_data, label='HF', color='darkcyan', linestyle="--")
plt.plot(r_array, fci_data, label='FCI', color='chocolate')
plt.plot(r_array, standard_vqe_data, label='VQE UCCSD', linestyle="--", color="crimson")
plt.plot(r_array, adapt_vqe_data_4_e3, label='ADAPT-VQE', marker='x', markersize=7, color='navy')

plt.ylabel("$E_{GS}$  (Hartree)")
plt.xlabel("R(Li-H) ($\mathrm{\AA}$)")
plt.legend()

plt.show()
```

```python
hf_data = abs(hf_data - fci_data)*au_to_kcal_conversion
standard_vqe_data = abs(standard_vqe_data - fci_data)*au_to_kcal_conversion
adapt_vqe_data_2 = abs(adapt_vqe_data_2 - fci_data)*au_to_kcal_conversion
adapt_vqe_data_4 = abs(adapt_vqe_data_4 - fci_data)*au_to_kcal_conversion
adapt_vqe_data_4_e3 = abs(adapt_vqe_data_4_e3 - fci_data)*au_to_kcal_conversion
```

```python
plt.plot(r_array, adapt_vqe_data_2, label='ADAPT-VQE 2', marker='o', markersize=5)
plt.plot(r_array, adapt_vqe_data_4, label='ADAPT-VQE 4', marker='o', markersize=5)
plt.plot(r_array, standard_vqe_data, label='VQE UCCSD')
plt.plot(r_array, hf_data, label='HF')
plt.plot(r_array, adapt_vqe_data_4_e3, label='ADAPT-VQE 4 3', marker='o', markersize=5)
plt.yscale('log')
plt.ylim(1e-10, 1e5)
plt.xlabel("R(H-H), Angstorm")
plt.ylabel("E-$E_{FCI}, kcal/mol$")
plt.title("LiH Molecules")

plt.axhspan(ymin=1e-10, ymax=1e0, color='blue', alpha=0.2, label='Chemical Accuracy (1 kcal/mol)')
plt.legend()
plt.show()
```
