```python
def calc_hf_N2H4(i):
    mol = pyscf.M(
        # atom = f'Li 0 0 0; H 0 0 {r}',  # in Angstrom
        # atom = 'N 0.0 0.0 0.0; H 0.0 0.0 1.015264; H 0.978541 0.0 -0.270591; N -0.334639 0.680561 -0.254663; H -0.605017 1.230432 0.554843; H 0.373523, 1.230432, -0.731012',
        atom = coordinates[i],
        basis = 'sto-3g',
        # symmetry = True,
    )
    myhf = mol.RHF().run()
    return myhf.kernel()
```
