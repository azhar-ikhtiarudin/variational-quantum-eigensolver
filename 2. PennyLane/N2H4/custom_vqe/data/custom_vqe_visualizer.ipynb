```python
import numpy as np
import matplotlib.pyplot as plt
au_to_kcal_conversion = 627.509474
```

### Basis Set Benchmark for Active Space 2 2

```python
data_3_2_2_sto_3g = np.loadtxt('N2H4_3_2_2_basis-sto-3g.txt', delimiter=',')
data_3_2_2_6_31g = np.loadtxt('N2H4_3_2_2_basis-6-31g.txt', delimiter=',')
data_3_2_2_6_311g = np.loadtxt('N2H4_3_2_2_basis-6-311g.txt', delimiter=',')
data_3_2_2_cc_pvdz = np.loadtxt('N2H4_3_2_2_basis-cc-pvdz.txt', delimiter=',')
```

```python
plt.plot(data_3_2_2_sto_3g, label='sto-3g')
plt.plot(data_3_2_2_6_31g, label='6-31g')
plt.plot(data_3_2_2_6_311g, label='6-311g')


plt.title('N2H4 ADAPT-VQE Custom Optimizer')
plt.legend()
plt.show()
```
